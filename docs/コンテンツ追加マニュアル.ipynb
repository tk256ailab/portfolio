## 概要

このポートフォリオサイトでは、コンテンツ（プロジェクトとブログ）をJSONファイルで管理しています。

### データファイルの場所

```
portfolio/
├── data/
│   ├── projects.json  # プロジェクト一覧
│   └── blogs.json     # ブログ記事一覧
└── assets/
    └── images/
        ├── projects/  # プロジェクト画像
        └── blogs/     # ブログ画像
```

### 追加の流れ

1. **画像を追加** - 必要に応じて画像を所定のフォルダに配置
2. **JSONファイルを編集** - プロジェクトまたはブログのデータを追加
3. **動作確認** - ローカルサーバーで表示を確認
4. **デプロイ** - GitHub Pagesにアップロード

## プロジェクトの追加

### 1. プロジェクト画像の準備

プロジェクトのサムネイル画像を `assets/images/projects/` フォルダに配置します。

**推奨サイズ**: 1200x800px または 16:9 の比率

### 2. data/projects.json の編集

`data/projects.json` を開き、配列の**先頭**に新しいプロジェクトを追加します。

### プロジェクトのデータ形式

```json
{
  "id": 6,
  "title": "プロジェクト名（日本語）",
  "titleEn": "Project Name (English)",
  "description": "プロジェクトの説明（日本語）",
  "descriptionEn": "Project description in English",
  "image": "assets/images/projects/your-image.png",
  "github": "https://github.com/username/repo",
  "demo": "https://demo-url.com",
  "tags": ["Python", "React", "Machine Learning"],
  "category": "web"
}
```

### フィールドの説明

| フィールド | 必須 | 説明 | 例 |
|-----------|------|------|----|
| **id** | ✅ | プロジェクトID（既存の最大値+1） | `6` |
| **title** | ✅ | プロジェクト名（日本語） | `"VRM Viewer"` |
| **titleEn** | ✅ | プロジェクト名（英語） | `"VRM Viewer"` |
| **description** | ✅ | 説明（日本語、2-3文） | `"WebブラウザでVRMファイルを表示..."` |
| **descriptionEn** | ✅ | 説明（英語、2-3文） | `"A viewer application that..."` |
| **image** | ❌ | 画像パス（nullも可） | `"assets/images/projects/vrm.png"` |
| **github** | ❌ | GitHubリポジトリURL | `"https://github.com/user/repo"` |
| **demo** | ❌ | デモサイトURL | `"https://demo.com"` |
| **tags** | ✅ | 技術タグ（配列） | `["JavaScript", "Three.js"]` |
| **category** | ✅ | カテゴリ | `"web"`, `"ml"`, `"tool"`, `"ai"` |

### 実際の追加例

**編集前** (`data/projects.json`):

```json
[
  {
    "id": 5,
    "title": "AIMascotKit",
    ...
  },
  ...
]
```

**編集後** (新しいプロジェクトを先頭に追加):

```json
[
  {
    "id": 6,
    "title": "新しいプロジェクト",
    "titleEn": "New Project",
    "description": "このプロジェクトは○○を実現するために開発しました。",
    "descriptionEn": "This project was developed to achieve ○○.",
    "image": "assets/images/projects/new-project.png",
    "github": "https://github.com/tk256ailab/new-project",
    "demo": "https://tk256ailab.github.io/new-project/",
    "tags": ["Python", "FastAPI", "React"],
    "category": "web"
  },
  {
    "id": 5,
    "title": "AIMascotKit",
    ...
  },
  ...
]
```

### ⚠️ 注意点

1. **JSONの構文エラーに注意**
   - カンマ `,` の位置に注意（最後の要素の後にはカンマ不要）
   - ダブルクォート `"` を使用（シングルクォート `'` は不可）
   - 画像パスは `assets/images/projects/` から始める（`../` は不要）

2. **IDは重複しないように**
   - 既存の最大ID + 1 を使用

3. **githubやdemoがない場合**
   - `null` を設定
   - 例: `"github": null`

## ブログ記事の追加

### 1. ブログ画像の準備（オプション）

サムネイル画像がある場合は `assets/images/blogs/` フォルダに配置します。

**推奨サイズ**: 1200x630px

### 2. data/blogs.json の編集

`data/blogs.json` を開き、配列の**先頭**に新しいブログ記事を追加します。

### ブログのデータ形式

```json
{
  "id": 14,
  "category": "ai",
  "noteId": "n1234567890ab",
  "platform": "note",
  "date": "2025.10.15",
  "image": "assets/images/blogs/my-article.webp",
  "title": "記事タイトル（日本語）",
  "titleEn": "Article Title (English)",
  "description": "記事の説明（日本語）",
  "descriptionEn": "Article description in English",

```

### フィールドの説明

| フィールド | 必須 | 説明 | 例 |
|-----------|------|------|----|
| **id** | ✅ | 記事ID（既存の最大値+1） | `14` |
| **category** | ✅ | カテゴリ | `"ai"`, `"web"`, `"algorithm"`, `"self-developed"`, `"blender"` |

| **date** | ✅ | 公開日（YYYY.MM.DD） | `"2025.10.15"` |
| **image** | ❌ | 画像パス（nullも可） | `"assets/images/blogs/article.webp"` |
| **title** | ✅ | タイトル（日本語） | `"機械学習入門"` |
| **titleEn** | ✅ | タイトル（英語） | `"ML Introduction"` |
| **description** | ✅ | 説明（日本語、2-3文） | `"TensorFlowの基本的な使い方..."` |
| **descriptionEn** | ✅ | 説明（英語、2-3文） | `"From basic TensorFlow usage..."` |
| **url** | ✅ | 記事URL | `"https://qiita.com/user/items/abc"` |

### ⚠️ 注意点

1. **プラットフォームに応じたIDを設定**
   - noteの場合: `noteId` を設定、`qiitaId` は不要

2. **noteのIDの取得方法**

3. **QiitaのIDの取得方法**
   - 記事URL: `https://qiita.com/username/items/abc123def456`

4. **カテゴリは既存のものから選択**
   - `ai`: AI・機械学習

## 画像の追加

### プロジェクト画像

```bash
# macOS/Linuxの場合
cp ~/Downloads/my-project-image.png assets/images/projects/

# Windowsの場合（PowerShell）
Copy-Item C:\Users\username\Downloads\my-project-image.png assets\images\projects\
```

### ブログ画像

```bash
# macOS/Linuxの場合
cp ~/Downloads/my-blog-image.webp assets/images/blogs/

# Windowsの場合（PowerShell）
Copy-Item C:\Users\username\Downloads\my-blog-image.webp assets\images\blogs\
```

### 推奨画像形式

- **プロジェクト**: PNG, JPG (推奨サイズ: 1200x800px)
- **ブログ**: WebP, PNG, JPG (推奨サイズ: 1200x630px)

### ファイル名の命名規則

- 小文字とハイフン `-` を使用
- 例: `vrm-viewer.png`, `ai-mascot-kit.webp`

## 動作確認

JSONファイルを編集したら、必ずローカルサーバーで動作確認を行ってください。

### 1. ローカルサーバーの起動

ターミナル（コマンドプロンプト）でポートフォリオディレクトリに移動し、以下のコマンドを実行します：

````python
# macOS/Linuxの場合
python3 -m http.server 8000

# Windowsの場合
python -m http.server 8000
````

### 2. ブラウザで確認

ブラウザで以下のURLにアクセスして確認します：

- **トップページ**: http://localhost:8000/index.html
- **プロジェクトページ**: http://localhost:8000/pages/projects.html
- **ブログページ**: http://localhost:8000/pages/blogs.html

### 3. 確認項目

✅ **プロジェクトページ**
- 新しいプロジェクトが表示されているか
- 画像が正しく表示されているか
- GitHubリンクとデモリンクが動作するか
- タグフィルターが動作するか
- 言語切り替え（JP/EN）が動作するか

✅ **ブログページ**
- 新しい記事が表示されているか
- 画像が正しく表示されているか（ある場合）
- 記事リンクが動作するか
- カテゴリフィルターが動作するか
- 言語切り替え（JP/EN）が動作するか

### 4. エラーが出た場合

ブラウザの開発者ツール（F12キー）でコンソールを確認してください。

**よくあるエラー**:

1. **JSON構文エラー**
   ```
   SyntaxError: Unexpected token } in JSON at position 123
   ```
   → カンマやクォートの位置を確認

2. **画像が表示されない**
   ```
   GET http://localhost:8000/assets/images/projects/xxx.png 404 (Not Found)
   ```
   → 画像パスとファイル名を確認

3. **データが表示されない**
   → ブラウザをリロード（Command+R / Ctrl+R）してキャッシュをクリア

## GitHub Pagesへのデプロイ

動作確認が完了したら、GitHub Pagesにアップロードします。

### 1. Gitリポジトリの準備（初回のみ）

まだGitリポジトリを作成していない場合：

````python
# ポートフォリオディレクトリで実行
git init
git add .
git commit -m "Initial commit"

# GitHubでリポジトリを作成してから
git remote add origin https://github.com/username/portfolio.git
git branch -M main
git push -u origin main
````

### 2. 変更をコミット＆プッシュ

````python
# 変更を確認
git status

# 変更をステージング
git add data/projects.json data/blogs.json assets/images/

# コミット
git commit -m "Add new project and blog article"

# GitHubにプッシュ
git push origin main
````

### 3. GitHub Pagesの設定（初回のみ）

1. GitHubのリポジトリページにアクセス
2. **Settings** タブをクリック
3. 左メニューから **Pages** を選択
4. **Source** で `main` ブランチを選択
5. フォルダは `/ (root)` を選択
6. **Save** をクリック

数分後、`https://username.github.io/portfolio/` でサイトが公開されます。

### 4. 公開サイトの確認

ブラウザで以下のURLにアクセスして確認します：

- **トップページ**: https://username.github.io/portfolio/
- **プロジェクトページ**: https://username.github.io/portfolio/pages/projects.html
- **ブログページ**: https://username.github.io/portfolio/pages/blogs.html

## トラブルシューティング

### JSONファイルの検証ツール

JSONファイルの構文が正しいか確認する方法：

````python
# macOS/Linuxの場合
python3 -m json.tool data/projects.json
python3 -m json.tool data/blogs.json

# Windowsの場合
python -m json.tool data\projects.json
python -m json.tool data\blogs.json
````

エラーが出なければJSON構文は正しいです。

### よくある問題と解決方法

| 問題 | 原因 | 解決方法 |
|------|------|----------|
| データが表示されない | JSONファイルのパスが間違っている | ブラウザのコンソールでエラーを確認 |
| 画像が表示されない | 画像ファイル名またはパスが間違っている | ファイル名とパスを確認（大文字小文字も区別される） |
| レイアウトが崩れる | HTMLタグが閉じられていない | ブラウザの開発者ツールで確認 |
| 言語切り替えが動かない | スクリプトの読み込みエラー | コンソールでエラーメッセージを確認 |
| GitHub Pagesで404エラー | リポジトリ設定が間違っている | Settings > Pages で設定を確認 |
